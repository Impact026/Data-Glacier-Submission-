Now that we've successfully cleaned our data, we can merge the dataframes into a single dataframe

```python
master_data = (transaction_ID_df.merge(cab_data_df, how= 'inner')).merge(customerID_df, how='inner').merge(city_data_df, how= 'inner')
print(master_data)
```

```python
master_data.info()
```

```python
master_data = transaction_ID_df.merge(cab_data_df, on="Transaction ID", how="inner")
master_data = master_data.merge(city_data_df, on="City", how="inner")
master_data = master_data.merge(customerID_df, on="Customer ID", how="inner")

print(master_data.info())
print("\n", master_data.head())
print(master_data.columns)
```

```python
# Assuming 'Timestamp' column exists in master_data
master_data['Hour'] = pd.to_datetime(master_data['Date of Travel']).dt.hour

hourly_counts = master_data.groupby(["Company", "Hour"]).size().unstack()

hourly_counts.plot(kind="line", title="Hours for each cab")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Users")
plt.legend(title="Company")
plt.show()

print(hourly_counts.idxmax())
```

```python
master_data["Profit"] = abs(master_data["Price Charged"] - master_data["Cost of Trip"])
print(master_data["Profit"])
```

```python
master_data["Profit Per KM"] = master_data["Profit"]/master_data["KM Travelled"]
master_data.info()
```

```python
index = 0
for date in master_data['Date of Travel']:
    days_since_2016 = date - 42364
    if days_since_2016 >= 0 and days_since_2016 <= 335:
        master_data.at[index, 'Year'] = 2016
    elif days_since_2016 <= 700:
        master_data.at[index, 'Year'] = 2017
    else:
        master_data.at[index, 'Year'] = 2018
    index = index + 1
    
rides_count = master_data['Customer ID'].value_counts().rename_axis('Customer ID').reset_index(name='Number of Rides')
print(master_data.head())
```

```python
pink_2016 = master_data.loc[(master_data['Company'] == 'Pink Cab') & (master_data['Year'] == 2016)]
pink_profit_2016 = pink_2016["Profit"].mean()

yellow_2016 = master_data.loc[(master_data['Company'] == 'Yellow Cab') & (master_data['Year'] == 2016)]
yellow_profit_2016 = pink_2016["Profit"].mean()


pink_2017 = master_data.loc[(master_data['Company'] == 'Pink Cab') & (master_data['Year'] == 2017)]
pink_profit_2017 = pink_2017["Profit"].mean()

yellow_2017 = master_data.loc[(master_data['Company'] == 'Yellow Cab') & (master_data['Year'] == 2017)]
yellow_profit_2017 = pink_2017["Profit"].mean()


pink_2018 = master_data.loc[(master_data['Company'] == 'Pink Cab') & (master_data['Year'] == 2018)]
pink_profit_2018 = pink_2018["Profit"].mean()


yellow_2018 = master_data.loc[(master_data['Company'] == 'Yellow Cab') & (master_data['Year'] == 2018)]
yellow_profit_2018 = yellow_2018["Profit"].mean()


```

```python
profit_forcabs = pd.DataFrame({
    'Year': [2016, 2017, 2018, 2016, 2017, 2018],
    'Company': ['Pink Cab', 'Pink Cab', 'Pink Cab', 'Yellow Cab', 'Yellow Cab', 'Yellow Cab'],
    'Profit': [pink_profit_2016, pink_profit_2017, pink_profit_2018, yellow_profit_2016, yellow_profit_2017, yellow_profit_2018]
})

profit_forcabs.info()
profit_forcabs["Profit"]

new_table = pd.pivot_table(profit_forcabs, values='Profit', index='Year', columns='Company')


new_table.plot(kind="line", title='Profit for Two Companies over Three Years')
plt.xlabel('Year')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```
