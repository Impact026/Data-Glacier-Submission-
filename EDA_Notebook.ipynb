```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

We have to perform Data Cleaning before actually using the data

```python
cab_url = "https://github.com/DataGlacier/DataSets/raw/main/Cab_Data.csv"

cab_data_df = pd.read_csv(cab_url)

print(cab_data_df.head())
print(cab_data_df.columns)
print(cab_data_df.dtypes)
print("shape before feature engineering = ", cab_data_df.shape)
```

```python
#for col in cab_data_df.columns:
 #   print(f'{col}: {cab_data_df[col].unique()}')
```

```python
print(cab_data_df.describe())
cab_data_df.drop_duplicates()
cab_data_df.dropna()
print("\n Results after checking for missing data:")
print(cab_data_df.isnull().sum())
print("shape after feature engineering = ", cab_data_df.shape)
```

```python
city_url = "https://raw.githubusercontent.com/DataGlacier/DataSets/main/City.csv"

city_data_df = pd.read_csv(city_url)

print(city_data_df.head())
print(city_data_df.columns)
print(city_data_df.dtypes)
print("shape before feature engineering = ", city_data_df.shape)
```

```python
print(city_data_df.describe())
city_data_df.drop_duplicates()
city_data_df.dropna()
print("\n Results after checking for missing data:")
print(city_data_df.isnull().sum())
print("shape after feature engineering = ", city_data_df.shape)
```

```python
customerID_url = "https://raw.githubusercontent.com/DataGlacier/DataSets/main/Customer_ID.csv"

customerID_df = pd.read_csv(customerID_url)

#print(customerID_df)
print(customerID_df.head())
print(customerID_df.columns)
print(customerID_df.dtypes)
print("shape before feature engineering = ", customerID_df.shape)
```

```python
print(customerID_df.describe()) 
customerID_df.drop_duplicates()
customerID_df.dropna()
print("\n Results after checking for missing data:")
print(customerID_df.isnull().sum())
print("shape after feature engineering = ", customerID_df.shape)  
```

```python
transaction_id_url = "https://github.com/DataGlacier/DataSets/raw/main/Transaction_ID.csv"
transaction_ID_df = pd.read_csv(transaction_id_url)

#print(transaction_ID_df)
print(transaction_ID_df.head)
print(transaction_ID_df.columns)
print(transaction_ID_df.dtypes)
print("shape before feature engineering = ",transaction_ID_df.shape)
```

```python
#for col2 in transaction_ID_df.columns:
    #print(f"{col2}: {transaction_ID_df[col2].unique()}")
```

```python
print(transaction_ID_df.describe())
transaction_ID_df.drop_duplicates()
transaction_ID_df.dropna()
print("\n Results after checking for missing data:")
print(transaction_ID_df.isnull().sum())
print("shape after feature engineering = ",transaction_ID_df.shape)
```

Now that we've successfully cleaned our data, we can merge the dataframes into a single dataframe

```python
master_data = (transaction_ID_df.merge(cab_data_df, how= 'inner')).merge(customerID_df, how='inner').merge(city_data_df, how= 'inner')
print(master_data)
```

```python
master_data.info()
```

```python
master_data = transaction_ID_df.merge(cab_data_df, on="Transaction ID", how="inner")
master_data = master_data.merge(city_data_df, on="City", how="inner")
master_data = master_data.merge(customerID_df, on="Customer ID", how="inner")

print(master_data.info())
print("\n", master_data.head())
print(master_data.columns)
```

```python
# Plot the distribution of cab users by city
master_data.groupby('City')['Customer ID'].count().plot(kind='bar', title='Cab Users by City')
plt.show()
```

```python
master_data.info()
```

```python
master_data["Company"].unique()
```

```python
cab_counts = master_data.groupby("Company")["Customer ID"].count()

cab_counts.plot(kind="bar", title="Number of Yellow Cabs to that of Pink Cabs")
plt.ylabel("Number of Rides")
plt.show()
```
